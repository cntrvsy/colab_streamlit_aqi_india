```python
def missing_values_table(combined_df):
    # Total missing values
    mis_val = combined_df.isnull().sum()

    # Percentage of missing values
    mis_val_percent = 100 * mis_val / len(combined_df)

    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    mis_val_table = mis_val_table.rename(columns={0: 'Missing Values', 1: '% of Total Values'})

    # Sort the table by percentage of missing descending
    mis_val_table = mis_val_table.sort_values('% of Total Values', ascending=False)

    return mis_val_table

missing_values = missing_values_table(combined_df)
display(missing_values.style.background_gradient(cmap='Reds'))
```

#DATA CLEANING

```python
combined_df['Date'] = pd.to_datetime(combined_df['Date'], format="mixed", errors="coerce")

# adding a month column
combined_df['Month'] = combined_df['Date'].dt.month

# adding a seasons column
combined_df['Season'] = combined_df['Month'].apply(lambda x: 'Winter' if x in [12, 1, 2] else (
    'Spring' if x in [3, 4, 5] else (
    'Summer' if x in [6, 7, 8] else (
    'Autumn' if x in [9, 10, 11] else 'Unknown'))))
```

```python
# Dealing with missing values
# Xylene
xylene_median = combined_df['Xylene'].median()
combined_df['Xylene'] = combined_df['Xylene'].fillna(xylene_median)
print(f"Missing values in 'Xylene' column after median imputation: {combined_df['Xylene'].isnull().sum()}")

#PM10
pm10_median = combined_df['PM10'].median()
combined_df['PM10'] = combined_df['PM10'].fillna(pm10_median)
print(f"Missing values in 'PM10' column after median imputation: {combined_df['PM10'].isnull().sum()}")

#NH3
nh3_median = combined_df['NH3'].median()
combined_df['NH3'] = combined_df['NH3'].fillna(nh3_median)

#Toluene
toluene_median = combined_df['Toluene'].median()
combined_df['Toluene'] = combined_df['Toluene'].fillna(toluene_median)

#Benzene
benzene_median = combined_df['Benzene'].median()
combined_df['Benzene'] = combined_df['Benzene'].fillna(benzene_median)

#AQI
aqi_median = combined_df['AQI'].median()
combined_df['AQI'] = combined_df['AQI'].fillna(aqi_median)

#PM2.5
pm25_median = combined_df['PM2.5'].median()
combined_df['PM2.5'] = combined_df['PM2.5'].fillna(pm25_median)

#NOx
nox_median = combined_df['NOx'].median()
combined_df['NOx'] = combined_df['NOx'].fillna(nox_median)

#O3
o3_median = combined_df['O3'].median()
combined_df['O3'] = combined_df['O3'].fillna(o3_median)

#SO2
so2_median = combined_df['SO2'].median()
combined_df['SO2'] = combined_df['SO2'].fillna(so2_median)

#NO2
no2_median = combined_df['NO2'].median()
combined_df['NO2'] = combined_df['NO2'].fillna(no2_median)

#NO
no_median = combined_df['NO'].median()
combined_df['NO'] = combined_df['NO'].fillna(no_median)

#CO
co_median = combined_df['CO'].median()
combined_df['CO'] = combined_df['CO'].fillna(co_median)
```

```python
def aqi_to_bucket(aqi):
    if pd.isna(aqi):
        return np.nan
    aqi = float(aqi)
    if aqi <= 50:
        return "good"
    if aqi <= 100:
        return "satisfactory"
    if aqi <= 200:
        return "moderate"
    if aqi <= 300:
        return "poor"
    if aqi <= 400:
        return "very poor"
    return "severe"

combined_df['AQI_Bucket'] = combined_df['AQI'].apply(aqi_to_bucket)

# Display the first few rows to show the updated AQI_Bucket
display(combined_df[['AQI', 'AQI_Bucket']].head())
```

PM2.5 is a key pollutant primarily because of its significant impact on human health and the environment.
- **Size Matters for Health:** PM2.5 refers to particulate matter with a diameter of 2.5 micrometers or less. Because these particles are so small, they can bypass the body's natural defense mechanisms and penetrate deep into the lungs, and some can even enter the bloodstream.
- **Health Impacts:** Exposure to high levels of PM2.5 is linked to a wide range of serious health problems, including:
  - Respiratory issues (asthma, bronchitis, reduced lung function)
  - Cardiovascular diseases (heart attacks, strokes)
  - Premature death
  - Exacerbation of existing lung and heart conditions
- **Sources:** PM2.5 can come from various sources, both natural and man-made, such as vehicle exhaust, industrial emissions, burning fossil fuels, wildfires, and certain chemical reactions in the atmosphere.
- **Environmental Impacts:** Beyond health, PM2.5 also contributes to reduced visibility (haze), acid rain, and can affect ecosystems by altering nutrient balances in soil and water.
