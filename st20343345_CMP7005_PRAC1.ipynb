If PM2.5 is high, it might drive the AQI to 'Poor'. If O3 is high, it could do the same. So, it's not a sum, but rather a representation of the dominant pollutant's impact on health.

```python
def missing_values_table(combined_df):
    # Total missing values
    mis_val = combined_df.isnull().sum()

    # Percentage of missing values
    mis_val_percent = 100 * mis_val / len(combined_df)

    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    mis_val_table = mis_val_table.rename(columns={0: 'Missing Values', 1: '% of Total Values'})

    # Sort the table by percentage of missing descending
    mis_val_table = mis_val_table.sort_values('% of Total Values', ascending=False)

    return mis_val_table

missing_values = missing_values_table(combined_df)
display(missing_values.style.background_gradient(cmap='Reds'))
```

#DATA CLEANING
 - For most of the pollutant columns (PM2.5, NO, NO2, NOx, NH3, CO, SO2, O3, Benzene, Toluene, and AQI), the median is likely the most appropriate choice as Pollutant levels are rarely perfectly normally distributed and often have higher frequencies at lower concentrations with occasional high peaks.

 - we will add a "month" and "season" column as well for more data anylsis.

```python
combined_df['Date'] = pd.to_datetime(combined_df['Date'], format="mixed", errors="coerce")

# adding a month column
combined_df['Month'] = combined_df['Date'].dt.month

# adding a seasons column
combined_df['Season'] = combined_df['Month'].apply(lambda x: 'Winter' if x in [12, 1, 2] else (
    'Spring' if x in [3, 4, 5] else (
    'Summer' if x in [6, 7, 8] else (
    'Autumn' if x in [9, 10, 11] else 'Unknown'))))
```

```python
# Dealing with missing values
# Xylene
xylene_median = combined_df['Xylene'].median()
combined_df['Xylene'] = combined_df['Xylene'].fillna(xylene_median)

#PM10
pm10_median = combined_df['PM10'].median()
combined_df['PM10'] = combined_df['PM10'].fillna(pm10_median)

#NH3
nh3_median = combined_df['NH3'].median()
combined_df['NH3'] = combined_df['NH3'].fillna(nh3_median)

#Toluene
toluene_median = combined_df['Toluene'].median()
combined_df['Toluene'] = combined_df['Toluene'].fillna(toluene_median)

#Benzene
benzene_median = combined_df['Benzene'].median()
combined_df['Benzene'] = combined_df['Benzene'].fillna(benzene_median)

#AQI
aqi_median = combined_df['AQI'].median()
combined_df['AQI'] = combined_df['AQI'].fillna(aqi_median)

#PM2.5
pm25_median = combined_df['PM2.5'].median()
combined_df['PM2.5'] = combined_df['PM2.5'].fillna(pm25_median)

#NOx
nox_median = combined_df['NOx'].median()
combined_df['NOx'] = combined_df['NOx'].fillna(nox_median)

#O3
o3_median = combined_df['O3'].median()
combined_df['O3'] = combined_df['O3'].fillna(o3_median)

#SO2
so2_median = combined_df['SO2'].median()
combined_df['SO2'] = combined_df['SO2'].fillna(so2_median)

#NO2
no2_median = combined_df['NO2'].median()
combined_df['NO2'] = combined_df['NO2'].fillna(no2_median)

#NO
no_median = combined_df['NO'].median()
combined_df['NO'] = combined_df['NO'].fillna(no_median)

#CO
co_median = combined_df['CO'].median()
combined_df['CO'] = combined_df['CO'].fillna(co_median)
```

## dealing with AQI Bucket
- We need a function to calculate AQI Bucket. it should be noted that different countries calculate this differently but for india it is as so:

| AQI Range | Category     | Health Advisory                                                                 |
|-----------|--------------|----------------------------------------------------------------------------------|
| 0–50      | Good         | None                                                                             |
| 51–100    | Satisfactory | Unusually sensitive people should reduce prolonged or heavy exertion            |
| 101–200   | Moderate     | Sensitive groups should reduce prolonged or heavy exertion                      |
| 201–300   | Poor         | Sensitive groups should avoid all outdoor activity; general public should limit prolonged or heavy exertion |
| 301–400   | Very Poor    | Sensitive groups should avoid all outdoor activity; general public should limit outdoor activity |
| 401–500   | Severe       | Everyone should avoid all physical activity outdoors                            |


```python
def aqi_to_bucket(aqi):
    if pd.isna(aqi):
        return np.nan
    aqi = float(aqi)
    if aqi <= 50:
        return "good"
    if aqi <= 100:
        return "satisfactory"
    if aqi <= 200:
        return "moderate"
    if aqi <= 300:
        return "poor"
    if aqi <= 400:
        return "very poor"
    return "severe"

combined_df['AQI_Bucket'] = combined_df['AQI'].apply(aqi_to_bucket)
```

PM2.5 is a key pollutant primarily because of its significant impact on human health and the environment.
- **Size Matters for Health:** PM2.5 refers to particulate matter with a diameter of 2.5 micrometers or less. Because these particles are so small, they can bypass the body's natural defense mechanisms and penetrate deep into the lungs, and some can even enter the bloodstream.
- **Health Impacts:** Exposure to high levels of PM2.5 is linked to a wide range of serious health problems, including:
  - Respiratory issues (asthma, bronchitis, reduced lung function)
  - Cardiovascular diseases (heart attacks, strokes)
  - Premature death
  - Exacerbation of existing lung and heart conditions
- **Sources:** PM2.5 can come from various sources, both natural and man-made, such as vehicle exhaust, industrial emissions, burning fossil fuels, wildfires, and certain chemical reactions in the atmosphere.
- **Environmental Impacts:** Beyond health, PM2.5 also contributes to reduced visibility (haze), acid rain, and can affect ecosystems by altering nutrient balances in soil and water.

#ML MODELS
We will be using Random Forest Regressor since our dataset is tabular in nature. this way we can
- **Handles Non-linearity**: It can capture complex, non-linear relationships between pollutants and the target.
- **Robustness**: Less prone to overfitting than a single decision tree and generally robust to outliers.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

##---Prepare the data---
# We'll use the 'chart_df' as it has 'PM2_5' (renamed from PM2.5) and imputed values.
# Exclude 'City', 'Date', 'Month', 'Season', 'AQI_Bucket' as features for AQI prediction.

# Ensure all feature columns are numeric. 'AQI' is the target.
feature_columns = ['PM2_5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
target_column = 'AQI'

X = chart_df[feature_columns]
y = chart_df[target_column]

# Verify no null values in X or y, as they should have been handled during cleaning
# If there are still any NaN values, we'll fill them with the median or drop rows, depending on context.
# For this demonstration, we assume they are filled as per previous steps.

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

```python
# Initialize the Random Forest Regressor model
# n_estimators is the number of trees in the forest
# random_state for reproducibility
model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1) # n_jobs=-1 uses all available processors

# Train the model on the training data
print("Training Random Forest Regressor...")
model.fit(X_train, y_train)
print("Model training complete.")
```

```python
##--Make predictions on the test set--
y_pred = model.predict(X_test)

# Evaluate the model's performance
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\nModel Evaluation:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2): {r2:.2f}")

# You can also look at feature importance
feature_importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nFeature Importances:")
display(feature_importances)
```

```python
%%writefile modelling_prediction.py
import streamlit as st
import pandas as pd
import joblib
import os

# Define the directory where model artifacts are saved
MODEL_DIR = 'streamlit_artifacts'

# Load the trained model and encoder (using Streamlit's cache for efficiency)
@st.cache_resource
def load_model_artifacts():
    model_path = os.path.join(MODEL_DIR, 'random_forest_model.pkl')
    encoder_path = os.path.join(MODEL_DIR, 'one_hot_encoder.pkl')
    features_path = os.path.join(MODEL_DIR, 'model_features.pkl')

    model = joblib.load(model_path)
    encoder = joblib.load(encoder_path)
    model_features = joblib.load(features_path)
    return model, encoder, model_features

def app():
    st.title("Modelling and Prediction")

    # Load model and encoder
    model_extended, encoder, model_features = load_model_artifacts()

    st.subheader("Model Evaluation")
    st.markdown("Here's a summary of the extended model's performance:")
    st.table(pd.DataFrame({
        "Metric": ["Mean Absolute Error (MAE)", "Mean Squared Error (MSE)", "Root Mean Squared Error (RMSE)", "R-squared (R2)"],
        "Value": [f"{19.91:.2f}", f"{1767.63:.2f}", f"{42.04:.2f}", f"{0.89:.2f}"] # Using previously computed values
    }))
    st.info("The model with extended features (including Month and City) shows a slight improvement in performance compared to the pollutants-only model, explaining 89% of the variance in AQI.")

    st.subheader("Predict AQI for Custom Inputs")
    st.markdown("Adjust the pollutant levels, month, and city below to get an estimated AQI.")

    # Get unique cities from the encoder (assuming the encoder was fitted on the full dataset)
    all_cities = sorted(encoder.categories_[0].tolist())

    # Input widgets for prediction
    col1, col2, col3 = st.columns(3)
    with col1:
        st.write("**Pollutant Levels (μg/m³ unless specified)**")
        pm2_5 = st.slider('PM2.5', 0.0, 500.0, 55.0, 0.1)
        pm10 = st.slider('PM10', 0.0, 1000.0, 100.0, 0.1)
        no = st.slider('NO', 0.0, 400.0, 15.0, 0.1)
        no2 = st.slider('NO2', 0.0, 400.0, 30.0, 0.1)
        nox = st.slider('NOx', 0.0, 500.0, 40.0, 0.1)
    with col2:
        st.write("") # Spacer
        nh3 = st.slider('NH3', 0.0, 400.0, 20.0, 0.1)
        co = st.slider('CO (mg/m³)', 0.0, 200.0, 1.5, 0.1)
        so2 = st.slider('SO2', 0.0, 200.0, 10.0, 0.1)
        o3 = st.slider('O3', 0.0, 300.0, 35.0, 0.1)
        benzene = st.slider('Benzene', 0.0, 500.0, 2.0, 0.1)
        toluene = st.slider('Toluene', 0.0, 500.0, 5.0, 0.1)
    with col3:
        st.write("") # Spacer
        xylene = st.slider('Xylene', 0.0, 200.0, 1.0, 0.1)

        st.write("**Temporal & Location**")
        month = st.selectbox('Month', range(1, 13), index=0) # Jan = 1, Dec = 12
        city = st.selectbox('City', all_cities, index=all_cities.index('Ahmedabad') if 'Ahmedabad' in all_cities else 0)

    if st.button('Predict AQI'):
        # Create a DataFrame for the new data point, ensuring all columns match model_features
        new_data_point = pd.DataFrame(0.0, index=[0], columns=model_features)

        # Populate pollutant values
        new_data_point['PM2_5'] = pm2_5
        new_data_point['PM10'] = pm10
        new_data_point['NO'] = no
        new_data_point['NO2'] = no2
        new_data_point['NOx'] = nox
        new_data_point['NH3'] = nh3
        new_data_point['CO'] = co
        new_data_point['SO2'] = so2
        new_data_point['O3'] = o3
        new_data_point['Benzene'] = benzene
        new_data_point['Toluene'] = toluene
        new_data_point['Xylene'] = xylene

        # Populate month
        new_data_point['Month'] = month

        # Populate one-hot encoded city
        city_col_name = f'City_{city}'
        if city_col_name in new_data_point.columns:
            new_data_point[city_col_name] = 1.0
        else:
            st.error(f"Selected city '{city}' was not found in the model's training data features. Please select another city.")
            return

        # Make prediction
        try:
            predicted_aqi = model_extended.predict(new_data_point)[0]
            st.success(f"Predicted AQI for the given conditions: **{predicted_aqi:.2f}**")

            # Convert AQI to bucket for better interpretation
            def aqi_to_bucket(aqi_val):
                if aqi_val <= 50: return "Good"
                elif aqi_val <= 100: return "Satisfactory"
                elif aqi_val <= 200: return "Moderate"
                elif aqi_val <= 300: return "Poor"
                elif aqi_val <= 400: return "Very Poor"
                else: return "Severe"

            aqi_category = aqi_to_bucket(predicted_aqi)
            st.info(f"This corresponds to an AQI category of: **{aqi_category}**")
        except Exception as e:
            st.error(f"An error occurred during prediction: {e}")
```
